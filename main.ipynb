```python
import pandas
import matplotlib.pyplot as pyplot

csv_path = 'csv/NationalNames.csv'
df = pandas.read_csv(csv_path)
gender_count_df = df.groupby(['Year', 'Gender'])['Count'].sum().reset_index()
gender_count_df = gender_count_df.pivot(index = 'Year', columns = 'Gender', values = 'Count')
pyplot.xlabel('Year')
pyplot.ylabel('Total births per year')
pyplot.plot(gender_count_df['M'])
pyplot.plot(gender_count_df['F'])
```

```python
import pandas
import matplotlib.pyplot as pyplot

csv_path = 'csv/NationalNames.csv'
df = pandas.read_csv(csv_path)
pyplot.xlabel('Year')
pyplot.ylabel('Count')
john_df = df.loc[df['Name'] == 'John']
mary_df = df.loc[df['Name'] == 'Mary']
pyplot.plot(john_df['Year'], john_df['Count'])
pyplot.plot(mary_df['Year'], mary_df['Count'])
```

```python
import pandas
import matplotlib.pyplot as pyplot

csv_path = 'csv/NationalNames.csv'
df = pandas.read_csv(csv_path)
pyplot.xlabel('Year')
pyplot.ylabel('Count')
john_df = df.loc[
    (df['Name'] == 'John') &
    (df['Gender'] == 'F')]
mary_df = df.loc[
    (df['Name'] == 'Mary') &
    (df['Gender'] == 'M')]
pyplot.plot(john_df['Year'], john_df['Count'])
pyplot.plot(mary_df['Year'], mary_df['Count'])
```
